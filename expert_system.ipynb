```python
from experta import *

class KnowledgeBase(KnowledgeEngine):
    def __init__(self):
        super().__init__()
        #Меры доверия
        self.MDIphon = 0
        self.MDXiaomi = 0
        self.MDGooglePixel = 0
        self.MDHuawei = 0
        self.MDSamsung = 0
        #Меры недоверия
        self.MNDIphon = 0
        self.MNDXiaomi = 0
        self.MNDGooglePixel = 0
        self.MNDHuawei = 0
        self.MNDSamsung = 0
    @Rule(
        Fact(answ="1-1") | Fact(answ="2-4") | Fact(answ="3-3") | Fact(answ="4-3") \
                         | Fact(answ="5-3") | Fact(answ="6-4") | Fact(answ="7-3"))
    def __a(self):
        self.MDIphon = self.MDIphon + 0.4 * (1 - self.MDIphon)
        self.MNDXiaomi  = self.MNDXiaomi  + 0.2 * (1 - self.MNDXiaomi )
        self.MNDGooglePixel = self.MNDGooglePixel + 0.2 * (1 - self.MNDGooglePixel)
        self.MNDHuawei = self.MNDHuawei + 0.2 * (1 - self.MNDHuawei)
        self.MNDSamsung = self.MNDSamsung + 0.2 * (1 - self.MNDSamsung)

    @Rule(
        Fact(answ="1-2") | Fact(answ="2-1") | Fact(answ="3-1") | Fact(answ="4-1") \
                         | Fact(answ="5-4") | Fact(answ="6-2") | Fact(answ="7-4"))
    def __plusAndrWin(self):
        self.MDXiaomi = self.MDXiaomi + 0.4 * (1 - self.MDXiaomi)
        self.MNDGooglePixel = self.MNDGooglePixel + 0.1 * (1 - self.MNDGooglePixel)
        self.MNDIphon = self.MNDIphon + 0.2 * (1 - self.MNDIphon)
        self.MNDHuawei = self.MNDHuawei + 0.3 * (1 - self.MNDHuawei)
        self.MNDSamsung = self.MNDSamsung + 0.4 * (1 - self.MNDSamsung)


    @Rule(
        Fact(answ="1-4") | Fact(answ="2-3") | Fact(answ="3-2") | Fact(answ="4-2") \
                         | Fact(answ="5-2") | Fact(answ="6-3") | Fact(answ="7-2"))
    def __plusAndWin2(self):
        self.MDGooglePixel = self.MDGooglePixel + 0.5 * (1 - self.MDGooglePixel)
        self.MNDXiaomi  = self.MNDXiaomi  + 0.3 * (1 - self.MNDXiaomi )
        self.MNDIphon = self.MNDIphon + 0.4 * (1 - self.MNDIphon)
        self.MNDHuawei = self.MNDHuawei + 0.2 * (1 - self.MNDHuawei)
        self.MNDSamsung = self.MNDSamsung + 0.3 * (1 - self.MNDSamsung)

    @Rule(
        Fact(answ="1-3") | Fact(answ="2-2") | Fact(answ="3-4") | Fact(answ="4-4") \
                         | Fact(answ="5-1") | Fact(answ="6-1") | Fact(answ="7-1"))
    def __plusAndWin2(self):
        self.MDHuawei = self.MDHuawei + 0.5 * (1 - self.MDHuawei)
        self.MNDGooglePixel = self.MNDGooglePixel + 0.3 * (1 - self.MNDGooglePixel)
        self.MNDXiaomi  = self.MNDXiaomi  + 0.3 * (1 - self.MNDXiaomi )
        self.MNDIphon = self.MNDIphon + 0.4 * (1 - self.MNDIphon)
        self.MNDSamsung = self.MNDSamsung + 0.3 * (1 - self.MNDSamsung)

    @Rule(
        Fact(answ="1-1") | Fact(answ="2-4") | Fact(answ="3-1") | Fact(answ="4-4") \
                         | Fact(answ="5-1") | Fact(answ="6-2") | Fact(answ="7-3"))
    def __plusAndWin2(self):
        self.MDSamsung = self.MDSamsung + 0.5 * (1 - self.MDSamsung)
        self.MNDGooglePixel = self.MNDGooglePixel + 0.5 * (1 - self.MNDGooglePixel)
        self.MNDXiaomi  = self.MNDXiaomi  + 0.3 * (1 - self.MNDXiaomi )
        self.MNDIphon = self.MNDIphon + 0.4 * (1 - self.MNDIphon)
        self.MNDHuawei = self.MNDHuawei + 0.2 * (1 - self.MNDHuawei)


    @Rule(
        Fact(answ="1-2") | Fact(answ="2-3") | Fact(answ="3-4") | Fact(answ="4-3") \
                         | Fact(answ="5-3") | Fact(answ="6-3") | Fact(answ="7-3"))
    def __plusIOS(self):
        self.MDIphon = self.MDIphon + 0.3 * (1 - self.MDIphon)
        self.MDXiaomi = self.MDXiaomi + 0.5 * (1 - self.MDXiaomi)
        self.MDGooglePixel = self.MDGooglePixel + 0.4 * (1 - self.MDGooglePixel)
        self.MDHuawei = self.MDHuawei + 0.2 * (1 - self.MDHuawei)
        self.MDSamsung = self.MDSamsung + 0.3 * (1 - self.MDSamsung)
        self.MNDIphon = self.MNDIphon + 0.1 * (1 - self.MNDIphon)
        self.MNDXiaomi  = self.MNDXiaomi  + 0.1 * (1 - self.MNDXiaomi )
        self.MNDGooglePixel = self.MNDGooglePixel + 0.1 * (1 - self.MNDGooglePixel)
        self.MNDHuawei = self.MNDHuawei + 0.1 * (1 - self.MNDHuawei)
        self.MNDSamsung = self.MNDSamsung + 0.1 * (1 - self.MNDSamsung)



    #Вспомогательные методы
    def getKU(self):
        return {
            "Xiaomi": self.MDXiaomi - self.MNDXiaomi ,
            "GooglePixel": self.MDGooglePixel - self.MNDGooglePixel,
            "Apple": self.MDIphon - self.MNDIphon,
            "Huawei": self.MDHuawei - self.MNDHuawei,
            "Samsung": self.MDSamsung - self.MNDSamsung
        }

    def checkPhones(self):
        phones = self.getKU()
        max = -1
        key = ""
        for ph in phones:
            if phones[ph] > max:
                max = phones[ph]
                key = ph

        return key, max
```

```python
KBase = KnowledgeBase()
KBase.reset() # Prepare the engine for the execution.

for ThisKey in Questions:
    print(Questions[ThisKey])
    x = str(input(Answers[ThisKey]))
    ans = str(ThisKey) + "-" + x
    KBase.declare(Fact(answ=ans))
    KBase.run()

check = KBase.checkPhones()

print("Вам подходят телефоны: " + check[0])
```
